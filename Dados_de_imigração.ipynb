## Analisar as tendencias de imigraÃ§Ã£o do Brasil em um periodo

```python
#Colocar o index como os paises do nosso dataframe
df.set_index('Country', inplace=True)
```

```python
anos = list(map(str,range(1980,2014)))
```

```python
anos
```

```python
brasil = df.loc['Brazil',anos]
```

```python
brasil
```

```python
#Vai criar um dicionario contendo o nome da coluna que eu quero e o dado de representaÃ§Ã£o.
brasil_dict = {'ano':brasil.index.tolist(),'imigrantes':brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict) #Transforma meu dicionario em um dataframe
```

```python
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'],dados_brasil['imigrantes'])
plt.title('ImigraÃ§Ã£o do Brasil para o CanadÃ¡')
plt.xlabel('Ano')
plt.ylabel('NÃºmero de imigrantes')
#para melhorar a visualizaÃ§Ã£o vamos definir as marcaÃ§Ãµes do eixo x de 5 em 5 anos
#Para isso vamos usar a funÃ§Ã£o xticks
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8,4))
ax.plot(dados_brasil['ano'],dados_brasil['imigrantes'],lw=3,color='g')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('ImigraÃ§Ã£o do Brasil para o CanadÃ¡\n1980 Ã¡ 2013',fontsize = 18,loc='left')
ax.set_xlabel('Ano',fontsize = 14)
ax.set_ylabel('NÃºmero de imigrantes',fontsize = 14)
ax.xaxis.set_tick_params(labelsize = 12)
ax.yaxis.set_tick_params(labelsize = 12)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
fig.savefig('imigraÃ§Ã£o_brasil_canadÃ¡.png',transparent=False,dpi=300, bbox_inches='tight')
plt.show()
```

```python
America_do_sul = df.query('Region == "South America"')
America_do_sul
```

```python
americaa_sul_sorted = America_do_sul.sort_values('Total',ascending=True)
#cores = ['royalblue','orange','forestgreen','orchid',
         #'purple','brown','slateblue','gray','olive','navy','teal','tomato']
```

```python
cores = []

for pais in americaa_sul_sorted.index:
  if pais == 'Brazil':
    cores.append('green')
  else:
    cores.append('silver')

fig, ax = plt.subplots(figsize = (14,5))
ax.barh(americaa_sul_sorted.index, americaa_sul_sorted['Total'], color = cores)
ax.set_title('America do sul: O Brasil Ã© o quarto pais com maior nÃºmero de imigrantes \n para o Canada entre 1980 e 2013',fontsize = 18, loc='left')
ax.set_xlabel('Numero de imigrantes',fontsize = 14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize = 12)
ax.yaxis.set_tick_params(labelsize = 12)

for i, v in enumerate(americaa_sul_sorted['Total']):
  ax.text(v + 0.2, i, str(v), color='black', fontweight='bold', fontsize = 10)

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both',which='both',length=0)
fig.savefig('ImigraÃ§Ã£o_america_do_sul.png',transparent=False,dpi=300, bbox_inches='tight')
plt.show()
```

```python
def gerar_grafico_paleta(palette):
  fig, ax = plt.subplots(figsize = (8,4))
  ax = sns.barplot(data=top_dez,y=top_dez.index, x='Total',orient='h',palette=palette)
  ax.set_title('Paises com maior imigraÃ§Ã£o para o CanadÃ¡\n1980 Ã¡ 2013',fontsize = 16, loc='left')
  ax.set_xlabel('NÃºmero de imigrantes',fontsize = 14)
  ax.set_ylabel('')

  plt.show()
```

```python
import plotly.graph_objs as go

# Criando uma figura
fig = go.Figure()

# Adicionando a linha do grÃ¡fico e definindo a espessura da linha
fig.add_trace(
    go.Scatter(x=[dados_brasil['ano'].iloc[0]], y=[dados_brasil['imigrantes'].iloc[0]], mode='lines', name='Imigrantes', line=dict(width=4))
)

# Definindo as configuraÃ§Ãµes de layout
fig.update_layout(
    title=dict(
        text='<b>ImigraÃ§Ã£o do Brasil para o CanadÃ¡ no perÃ­odo de 1980 a 2013</b>',
        x=0.12,
        xanchor='left',
        font=dict(size=20)
    ),
    xaxis=dict(range=[1980, 2013], autorange=False, title='<b>Ano</b>'),
    yaxis=dict(range=[0, 3000], autorange=False, title='<b>NÃºmero de imigrantes</b>'),
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]
        )]
    )],
    width=1000,
    height=500
)

# Definir as configuraÃ§Ãµes de animaÃ§Ã£o
frames = [go.Frame(data=[go.Scatter(x=dados_brasil['ano'].iloc[:i+1], y=dados_brasil['imigrantes'].iloc[:i+1])]) for i in range(len(dados_brasil))]
fig.frames = frames

# Mostrando a figura
fig.show()
```

```python
df_argentina['ano'] = df_argentina['ano'].astype(int)
```

```python
import plotly.graph_objs as go

# Criando uma figura
fig = go.Figure()

# Adicionando a linha com os dados do Brasil
fig.add_trace(
    go.Scatter(x=[dados_brasil['ano'].iloc[0]], y=[dados_brasil['imigrantes'].iloc[0]], mode='lines', name='Imigrantes do Brasil', line=dict(width=4))
)

# Adicionando a linha com os dados da Argentina
fig.add_trace(
    go.Scatter(x=[df_argentina['ano'].iloc[0]], y=[df_argentina['imigrantes'].iloc[0]], mode='lines', name='Imigrantes da Argentina', line=dict(width=4))
)

# Definindo as configuraÃ§Ãµes de layout
fig.update_layout(
    title=dict(
        text='<b>ImigraÃ§Ã£o do Brasil e da Argentina para o CanadÃ¡ no perÃ­odo de 1980 a 2013',
        x=0.1,

        font=dict(size=18)
    ),
    xaxis=dict(range=[1980, 2013], autorange=False, title='<b>Ano</b>'),
    yaxis=dict(range=[0, 3000], autorange=False, title='<b>NÃºmero de imigrantes</b>'),
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]
        )]
    )],
    width=1200, # largura da figura em pixels
    height=600 # altura da figura em pixels
)

# Definindo as configuraÃ§Ãµes de animaÃ§Ã£o
frames = []
for i in range(len(dados_brasil)):
    frame_data = [
        go.Scatter(x=dados_brasil['ano'].iloc[:i+1], y=dados_brasil['imigrantes'].iloc[:i+1]),
        go.Scatter(x=df_argentina['ano'].iloc[:i+1], y=df_argentina['imigrantes'].iloc[:i+1])
    ]
    frame = go.Frame(data=frame_data)
    frames.append(frame)
fig.frames = frames

# Mostrando a figura
fig.show()
```
