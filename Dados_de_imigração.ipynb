## Analisar as tendencias de imigraÃ§Ã£o do Brasil em um periodo

```python
#Colocar o index como os paises do nosso dataframe
df.set_index('Country', inplace=True)
```

```python
anos = list(map(str,range(1980,2014)))
```

```python
anos
```

```python
brasil = df.loc['Brazil',anos]
```

```python
brasil
```

```python
#Vai criar um dicionario contendo o nome da coluna que eu quero e o dado de representaÃ§Ã£o.
brasil_dict = {'ano':brasil.index.tolist(),'imigrantes':brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict) #Transforma meu dicionario em um dataframe
```

```python
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'],dados_brasil['imigrantes'])
plt.title('ImigraÃ§Ã£o do Brasil para o CanadÃ¡')
plt.xlabel('Ano')
plt.ylabel('NÃºmero de imigrantes')
#para melhorar a visualizaÃ§Ã£o vamos definir as marcaÃ§Ãµes do eixo x de 5 em 5 anos
#Para isso vamos usar a funÃ§Ã£o xticks
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8,4))
ax.plot(dados_brasil['ano'],dados_brasil['imigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('ImigraÃ§Ã£o do Brasil para o CanadÃ¡\n1980 Ã¡ 2013')
ax.set_xlabel('Ano')
ax.set_ylabel('NÃºmero de imigrantes')
plt.show()
```

```python
fig, axs = plt.subplots(1,2, figsize = (12,4))
axs[0].plot(dados_brasil['ano'],dados_brasil['imigrantes'])
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].set_title('ImigraÃ§Ã£o do Brasil para o CanadÃ¡\n1980 Ã¡ 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('NÃºmero de imigrantes')
axs[0].grid()

axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot da migraÃ§Ã£o do Brasil para o CanadÃ¡')
axs[1].set_ylabel('NÃºmero de imigrantes')
axs[1].grid()
plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize = (10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)

axs[0,0].plot(df.loc['Brazil',anos])
axs[0,0].set_title('Brasil')


axs[0,1].plot(df.loc['Colombia',anos])
axs[0,1].set_title('ColÃ´mbia')


axs[1,0].plot(df.loc['Argentina',anos])
axs[1,0].set_title('Argentina')


axs[1,1].plot(df.loc['Peru',anos])
axs[1,1].set_title('Peru')


for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat:
  ax.set_xlabel('Ano')
  ax.set_ylabel('NÃºmero de imigrantes')
  ax.grid()

plt.show()
```

## Desafio paralelo
criar um grÃ¡fico de linhas comparando os nÃºmeros de imigrantes do Brasil e Argentina para o CanadÃ¡, que sÃ£o os maiores paÃ­ses da AmÃ©rica do Sul.

```python
argentina = df.loc['Argentina', anos]
argentina
```

```python
argentina_dict = {'ano':argentina.index.tolist(),'imigrantes':argentina.values.tolist()}
df_argentina = pd.DataFrame(argentina_dict)
```

```python
df_argentina
```

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'],dados_brasil['imigrantes'])
plt.plot(df_argentina['ano'],df_argentina['imigrantes'])
plt.legend(['Brasil','Argentina'])
plt.title('ImigraÃ§Ã£o do Brasil e Argentina para o CanadÃ¡')
plt.xlabel('Ano')
plt.ylabel('NÃºmero de imigrantes')
#para melhorar a visualizaÃ§Ã£o vamos definir as marcaÃ§Ãµes do eixo x de 5 em 5 anos
#Para isso vamos usar a funÃ§Ã£o xticks
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.show()
```
